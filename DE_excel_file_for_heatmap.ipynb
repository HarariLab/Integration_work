```r
# load necessary packages
library(openxlsx)
library(gtools)
library(dplyr)
setwd("/home/a.garg/integration_project/step_15/before_annotation/part_1/")
```

```r
all_files <- list.files(pattern = "^microglia*") %>% mixedsort()
wb <- createWorkbook()
for(file in all_files) {
    df <- read.xlsx(file)
   df <-  df %>% relocate(gene, .before=logFC_cluster)
    addWorksheet(wb,file)
    writeData(wb,file,df)
    sheet_name <- gsub("^microglia", "", gsub(".xlsx", "", file), ignore.case = TRUE)
    renameWorksheet(wb, file, sheet_name)
}
saveWorkbook(wb,'/home/a.garg/integration_project/step_15/Microglia_nebula_ouput2.xlsx',overwrite=TRUE)
```
