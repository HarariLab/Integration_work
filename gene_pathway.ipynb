```r
library(openxlsx)
library(enrichR)
library(stringr)
library(gtools)
library(dplyr)

setwd("~/integration_project/step_16/part_1/")
```

```r
files = system('ls /home/a.garg/integration_project/step_15/before_annotation/part_1/microglia*xlsx',intern=T)
# read and save into excel
pages = lapply(files,function(f){
      name<- gsub(".*/ *", "\\1\\2",f)
})
pages 
Hits = data.frame(Term = NA)
```

```r
tmp = lapply(pages,function(p){
d = read.xlsx('/home/a.garg/integration_project/step_15/before_annotation/part_1/Microglia_nebula_ouput.xlsx',sheet=p)
 d = d[d$logFC_cluster > 0.25,]
 if(nrow(d) >= 2){
  r = enrichr(d$gene,c('KEGG_2021_Human','GO_Biological_Process_2021'))
  r = rbind(r$KEGG_2021_Human, r$GO_Biological_Process_2021)
  r[,str_remove(p, ".xlsx")] = -log10(r$P.value)
   write.xlsx(r, file = paste0("pathway_",p), sheetName = p, append = FALSE)
 }
})
```

```r
all_files <- list.files(pattern = "^pathway_*") %>% mixedsort()
wb <- createWorkbook()
for(file in all_files) {
    df <- read.xlsx(file)
    addWorksheet(wb,file)
    writeData(wb,file,df)
}
saveWorkbook(wb,'/home/a.garg/integration_project/step_16/part_1/Microglia_pathway_ouput.xlsx',overwrite=TRUE)
```
